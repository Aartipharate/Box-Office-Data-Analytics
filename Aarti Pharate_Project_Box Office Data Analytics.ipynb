## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('max_columns', None)
%matplotlib inline
plt.style.use('ggplot')
```

## Task 1: Data Loading and Exploration

```python
df = pd.read_csv("C:\\Users\\Hp\\Desktop\\Projects\\box_office.csv")
```

```python
df.head()
```

## Task 2: Visualizing the Target Distribution

```python
fig,axes = plt.subplots(figsize = (20, 7))
plt.subplot(1, 2, 1)
sns.distplot(df['revenue'], kde = False)
plt.title("Revenue Distribution")
plt.subplot(1, 2, 2)
sns.distplot(np.log1p(df['revenue']), kde = False)
plt.title("Log Revenue Distribution")
```

```python
fig, axes = plt.subplots(figsize = (20, 7))
plt.subplot(1, 2, 1)
sns.distplot(df['budget'], kde = False)
plt.title("Budget Distribution")
plt.subplot(1, 2, 2)
sns.distplot(np.log1p(df['budget']), kde = False)
plt.title("Log Budget Distribution")
```

## Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize = (20, 10))
plt.subplot(1, 2, 1)
sns.scatterplot(df['budget'], df['revenue'])
plt.title('Revenue vs Budget');
plt.subplot(1, 2, 2)
df['log_revenue'] = np.log1p(df['revenue'])
sns.scatterplot(np.log1p(df['budget']), df['log_revenue'])
plt.title('Log Revenue vs Budget')
```

```python
df['log_revenue'] = np.log1p(df['revenue'])
df['log_budget'] = np.log1p(df['budget'])
```

## Task 4: Distribution of Languages in Film

```python
language_data = df.loc[df['original_language'].isin(df['original_language'].value_counts().head().index)]
```

```python
plt.figure(figsize = (18, 9))
plt.subplot(1, 2, 1)
sns.boxplot(x = 'original_language', y = 'revenue', data = language_data)
plt.title('Mean Revenue per Language')
plt.subplot(1, 2, 2)
sns.boxplot(x = 'original_language', y = 'log_revenue', data = language_data)
plt.title('Mean Log Revenue per Language')
```
